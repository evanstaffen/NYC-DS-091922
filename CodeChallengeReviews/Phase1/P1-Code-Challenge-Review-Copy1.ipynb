## Task 4: Create a column

An "impression" counts each time a post is displayed.  

Create a new column called `likes_per_impression` which divides the number of likes per post by the number of impressions per post.

```python
# call_on_students(1)
```

```python
# Your code here, number of likes per post/# of impressions per post
#(total likes/total posts)/(total impressions/total post)
#16/7 / 8/7 

df['likes_per_post'] = df['like']/df['Lifetime Post Total Reach']
df['likes_per_post']

df['impressions_per_post'] = df['Lifetime Post Total Impressions'].astype(float)/df['Lifetime Post Total Reach']
df['impressions_per_post']

df['likes_per_impression'] = df['likes_per_post']/df['impressions_per_post']
df['likes_per_impression']
```
